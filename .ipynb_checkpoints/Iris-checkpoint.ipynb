# Iris Species Classification
## (setosa - versicolor - virginica )
This model predict the specie of an Iris based on spal and petal dimensions, using Logistic Regression Algorithm

### Tools and DataSet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
dataset = pd.read_csv('Iris_dataset.csv')
dataset.head()
```

```python
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Spliting the data into training and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Visualize the data

```python
x_features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(x_train[:,i], y_train)
    ax[i].set_xlabel(x_features[i])
ax[0].set_ylabel("Specie")
plt.show()
```

### Scaling the data using standardization

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

### Visualize the data after scaling

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(x_train[:,i], y_train)
    ax[i].set_xlabel(x_features[i])
ax[0].set_ylabel("Specie")
plt.show()
```

### Creating and training the logistic regression model

```python
model = LogisticRegression(multi_class = 'ovr', random_state = 0)
model.fit(x_train, y_train)
```

### Evaluation Metrics
Accuracy, Precision, Recall, F1 Score

```python
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-score: {f1}')
```
